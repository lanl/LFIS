```python
from LFIS.LF import LF_base
from LFIS.density import MG2D, MultiNormDist
import LFIS.schedule as schedule
import LFIS.model as model
from LFIS.util import  save_file, save_stat, run_stat, load_file
from LFIS.train import train_withweight, train_withoutweight
from copy import deepcopy
from torch.optim import Adam

import os

import torch
import numpy as np

import sys

torch.set_default_dtype(torch.float64)
```

```python
case_name = 'MG2D_256_Sigma1_Tsigma03_grid1_test'
print(case_name)

device =  'cuda:0'
ndim  = 2         ###  dimension of the distribution
task = 'train'     ###  train or test
```

```python
#################################                                                                                                       
mu_prior = torch.tensor([0.0, 0.0])  ### mean vector of the initial distribution
sigma2_prior = torch.eye(2)          ### covariance matrix of the initial distribution
cfl = 1.0
steps = 256                          #### number of steps
dtmax = torch.tensor(1.0/steps)      #### maximum stepsize
```

```python
### Setting up the initial distribution and target distribution                                                                         
prior = MultiNormDist(ndim,mu_prior, sigma2_prior, device = device )
target = MG2D(2, device=device, grid= 1, sigma2 = 0.3/25)


###### Setting up the schedule ##############                                                                                           
tt = schedule.CosSchedule()

##### Setting up the NN model to parametrize the Liouville flow #####                                                                   
flow_model  = model.NN_base(ndim,64,2)

#####  Assemble the LF model #########                                                                                                  
LFmodel = LF_base(2, flow_model, tt, target, prior )
LFmodel.to(device)
```

# Train modeltrain( LFmodel, lr = 0.01, cfl = None, dtmax = torch.tensor(0.01), epoch = 2000, nsample = 10000)       
### Default: Optimizer: Adam with initial lr = 0.01                                                                                     
###          Sample size: 10,000                                                                                                        
###          cfl: None, i.e. constant time step specificed by dtmax 

```python
if task == 'train':
    output = train_withweight(LFmodel, lr = 0.5e-3, cfl= cfl, dtmax = dtmax,nsample = 600000, nbatch=5000)
    save_file(case_name, output)


    logstat = run_stat(LFmodel, output)
    save_stat(case_name, logstat )

elif task == 'test':
    output = load_file(case_name, flow_model)
    logstat = run_stat(LFmodel, output, nruns=30, nsample=2000, withweight=True)
    save_stat(case_name, logstat , filename = 'LF_stats.npz')
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
def cal_pdf(x,n):
    ypdf,xpdf = np.histogram(x,n)
    xpdf = xpdf[:-1]+ (xpdf[1]-xpdf[0])/2
    ypdf = ypdf/(ypdf*(xpdf[1]-xpdf[0])).sum()
    return xpdf,ypdf
```

```python
MG2D_xm = np.linspace(-1.5,1.5,100)
MG2D_ym = np.linspace(-1.5,1.5,100)
Xmesh, Ymesh = np.meshgrid(MG2D_xm,MG2D_ym)
mesh = torch.from_numpy(np.stack([Xmesh,Ymesh],axis=2).reshape([-1,2]))

MG2D_contourgt = torch.exp(target.evaluate_log_density(mesh.to(device)))
```

```python
MG2D_pdf = torch.exp(MG2D_contourgt).reshape([100,100]).cpu().detach()
LFIS_samples = logstat['samples']
LFIS_weights = logstat['weight']
```

```python
fig,ax = plt.subplots(1,2,constrained_layout = True, figsize=  (8,3.5))
ax[0].contourf(MG2D_xm,MG2D_ym,MG2D_pdf,levels=100, cmap=cm.plasma)
ax[1].scatter(LFIS_samples[:,0], LFIS_samples[:,1],s = 0.5*LFIS_weights )
ax[0].set_title('GT PDF')
ax[1].set_title('LFIS samples')
```
